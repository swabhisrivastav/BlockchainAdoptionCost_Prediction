script generate a simple sales dataset with columns for Date, Product, Quantity Sold, and Revenue. Each product has sales data for 30 days

```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Generate sample data dictionary
data = {
    'Date': [datetime(2023, 11, 1) + timedelta(days=i) for i in range(30)],
    'Product': [f'Product_{i}' for i in range(1, 11) for _ in range(3)],
    'Quantity Sold': [random.randint(10, 100) for _ in range(30)],
}

# Calculate Revenue based on Quantity Sold and a random price
data['Revenue'] = [quantity * random.uniform(10, 50) for quantity in data['Quantity Sold']]

# Create a DataFrame
sales_df = pd.DataFrame(data)

sales_df.to_csv('sales_data.csv',index=False)
print(sales_df)
```
